## Checking glucose level distributions 

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Glucose'], kde=True, bins=20)
plt.title("Glucose Distribution")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.show()
```

* Glucose levels are not normally distributed

## Assigning upper and lower bounds to the data to determine normal glucose levels and identify outliers

```python

col = 'Glucose'

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")

# Identify outliers
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]

print(f"Number of outliers: {len(outliers)}")
print("Outlier values:")
print(outliers)
```

```python
# Creating a df with outliers removed
glucose_clean = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)][col]
print(f"Number of remaining values: {len(glucose_clean)}")
```

```python
# After removing outliers

plt.figure(figsize=(6,5))
sns.histplot(glucose_clean, kde=True, bins=20, color='green')
plt.title("Glucose - Outliers Removed")

plt.show()
```
